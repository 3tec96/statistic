## 2. 変数1個の場合の解析方法（value1のみ使用）

### 2.1. 判別方式

ある結果$value1 = x_1$を持つデータが陽性か陰性かを判定したい。

マハラノビス距離
$
{D_1}^2 = \frac{(x_1 - \mu_{11})^2}{\sigma^2} , {D_2}^2 = \frac{(x_2 - \mu_{12})^2}{\sigma^2}
$
のうち小さいほうに分類する。

次の$z$を線形判別関数といい、$z$の符号で判別を行う。

$
z = \frac{{D_2}^2 - {D_1}^2}{2} = \frac{(x_2 - \mu_2)^2 - (x_1 - \mu_1)^2}{2\sigma^2}
= \frac{(\mu_{11}+\mu_{12})}{\sigma^2} (x_1 - \frac{\mu_{11}+\mu_{12}}{2})
$

通常、母平均、母分散は未知なので、データからこれらを推定する。
各サンプルに対して求めた$z = \hat{z}$の値をスコア（判別得点）と呼ぶ。

```python
# detaをarrayに変換
t = df["result"].values 
x1 = df["value1"].values
x2 = df["value2"].values
```

```python
# t==1が陽性,t==0が陰性
```

```python
# データの個数
n1 = np.count_nonzero(t==1)
n2 = np.count_nonzero(t==0)
print("陽性の個数：", n1)
print("陰性の個数：", n2)
```

```python
# value1の平均値
mu11_hat = x1[np.where(t==1)].sum()/n1
mu12_hat = x1[np.where(t==0)].sum()/n2
print("陽性の人のvalue1の平均値：", mu11_hat)
print("陰性の人のvalue1の平均値：", mu12_hat)
```

```python
# value1の分散
var11_hat = x1[np.where(t==1)].var(ddof=1)
var12_hat = x1[np.where(t==0)].var(ddof=1)
print("陽性の人のvalue1の分散：", var11_hat)
print("陰性の人のvalue1の分散：", var12_hat)
```

```python
# 平均値とプールした分散を求める
mu1_hat = (mu11_hat + mu12_hat) / 2
var1_hat = ((n1 - 1) * var11_hat + (n2 - 1) * var12_hat) / (n1 + n2 - 2)
print("平均値：", mu1_hat)
print("分散　：", var1_hat)
```

```python
# これを代入すれば線形判別関数zを求めることができる。
a = (mu11_hat - mu12_hat) / var1_hat
b = -a * mu1_hat
print("傾き：", a)
print("切片：", b)
```

判別分析では、陽性、陰性の母集団が正規分布に従うことを仮定しているようだ。
そして二つの母集団の分散を等しいとしている。

### 2.2. 誤判別率

```python
# 誤判別率の推定
std = (mu11_hat - mu1_hat) / np.sqrt(var1_hat)
p = norm.cdf(std, loc=0, scale=1)
print("誤判別率：", 1 - p)
```

```python
# 学習データを使って正解率を確認
y_pred = ((a*x1 + b)>0).astype(int)
print("予測　：", y_pred)
print("正解　：", t)
print("正解率：", (y_pred==t).sum() / t.shape[0])
```

二つの母集団の平均の差が大きいほど誤判別率は低くなる。
そこで、母集団間の距離として2つの母集団の母平均のマハラノビス距離の2乗を判別効率よぶ。

```python
# 判別効率
print("判別効率：",(mu11_hat - mu12_hat)**2 / var1_hat)
```
